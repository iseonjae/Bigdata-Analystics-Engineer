1. '관광객비율'이 두 번째로 높은 국가의 '사업' 방문객 수를 a라고 정의하시오.
2. '관광'이 두 번째로 높은 국가의 '공무' 방문객 수를 b라고 정의하시오.
3. a와 b의 합을 구하시오.
- 방문객 합계 = 관광 + 공무 + 사업 + 기타
- 관광객 비율 = 관광 / 방문객 합계

```python
import pandas as pd
df = pd.read_csv('/Users/iseonjae/Desktop/빅데이터분석기사/8회기출/tourist.csv')
```

```python
print(df.shape)
print(df.head())
print(df.info())
```

```python
print(df.head())
```

```python
a = df.sort_values(by='관광객비율', ascending=False).head(2).iloc[1,3]
print(a)
```

```python
# 관광이 두 번째로 높은 나라의 공무 방문객 수 구하기
b = df.sort_values(by='관광', ascending=False).head(2).iloc[1,2]
print(b)
```

```python
print(a+b)
```

1. 주어진 데이터에서 'co'와 'nmhc' 컬럼을 각각 Min-Max 스케일링하시오.
2. 스케일링된 'co', 'rmhc' 컬럼의 표준편차를 각각 구하시오.
3. 'co' 컬럼의 표준편차에서 'nmhc' 컬럼의 표준편차를 뺀 값을 소수점 3자리로 반올림하여 구하시오.
- Min-Max 스케일링 = (X - min_X) / (Max_X - min_X)
- X 변수의 최솟값 : min_X
- X 변수의 최댓값 : max_X

```python
import pandas as pd
df = pd.read_csv('/Users/iseonjae/Desktop/빅데이터분석기사/8회기출/chem.csv')
```

```python
print(df.shape)
print(df.head())
print(df.info())
```

```python
# Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['co_scaled'] = scaler.fit_transform(df[['co']])
df['nmhc_scaled'] = scaler.fit_transform(df[['nmhc']])
```

```python
print(df.head())
```

```python
# 표준편차 구하기
r1 = df['co_scaled'].std()
r2 = df['nmhc_scaled'].std()
```

```python
print(r1)
print(r2)
```

```python
print(round(r1-r2, 3))
```

통신사에서 고객에게 청구될 총 금액을 예측하시오. (MAE)

```python
import pandas as pd
train = pd.read_csv('/Users/iseonjae/Desktop/빅데이터분석기사/8회기출/churn_train.csv')
test = pd.read_csv('/Users/iseonjae/Desktop/빅데이터분석기사/8회기출/churn_test.csv')
```

```python
# 탐색적 자료 분석
# 데이터 개수 파악
print(train.shape)
print(test.shape)
```

```python
# 데이터 미리보기 
print(train.head())
print(test.head())
```

```python
# 데이터 정보 파악
print(train.info())
print(test.info())
```

```python
# 결측치 개수 파악
print(train.isnull().sum().sum())
print(test.isnull().sum().sum())
```

```python
# 목표 변수 분포 확인
print(train['TotalCharges'].describe())
```

```python
# 데이터 전처리
# 목표 변수 분리
target = train.pop('TotalCharges')
```

```python
# ID 변수 삭제
train.drop('customerID', axis=1, inplace=True)
test.drop('customerID', axis=1, inplace=True)
```

```python
# 범주형 변수 원-핫 인코딩
combined = pd.concat([train, test])
combined_dummies = pd.get_dummies(combined)
n_train = len(train)
train = combined_dummies[:n_train]
test = combined_dummies[n_train:]
```

```python
# 검증 데이터 분할
from sklearn.model_selection import train_test_split
X_train, X_val, Y_train, Y_val = train_test_split(train, target, test_size=0.2, random_state=0)
```

```python
# 머신러닝 모델 학습
# 의사결정나무
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(max_depth=5, random_state=0)
dt.fit(X_train, Y_train)
pred = dt.predict(X_val)
```

```python
# 모델 평가
from sklearn.metrics import mean_absolute_error
score = mean_absolute_error(Y_val, pred)
print(score)
# 1256.96348907767
# 935.1231478759449 (max_depth=5)
```

```python
# xgboost
import xgboost as xgb
xg = xgb.XGBRegressor(random_state=0)
xg.fit(X_train, Y_train)
pred = xg.predict(X_val)
```

```python
score = mean_absolute_error(Y_val, pred)
print(score)
```

```python
# 랜덤 포레스트
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=5, n_estimators=500, random_state=0)
rf.fit(X_train, Y_train)
pred = rf.predict(X_val)
```

```python
score = mean_absolute_error(Y_val, pred)
print(score)
# 949.6177326752966
# 914.4499668658017 (max_depth=5)
# 914.2298332116615 (max_depth=5, n_estimators=300)
# 914.0195044213503 (max_depth=5, n_estimators=500)
```

```python
# 결과 예측 및 파일 생성
pred = rf.predict(test)
submit = pd.DataFrame({'pred':pred})
submit.to_csv('result.csv', index=False)
```

1. 주어진 데이터에서 로지스틱 회귀 분석을 수행해 유의확률이 0.05 이상인 유의하지 않은 독립변수의 개수를 구하시오. (종속변수 : Churn, 독립변수 : Churn을 제외한 모든 변수)

```python
import pandas as pd
df = pd.read_csv('/Users/iseonjae/Desktop/빅데이터분석기사/8회기출/churn.csv')
```

```python
print(df.shape)
print(df.head())
print(df.info())
```

```python
# 로지스틱 회귀
from statsmodels.formula.api import logit
model = logit('Churn ~ AccountWeeks + ContractRenewal + DataPlan + DataUsage + CustServCalls + DayMins + DayCalls + MonthlyCharge + OverageFee + RoamMins', data=df).fit()
print(model.summary())
```

```python
# 유의확률 0.05 이상인 독립변수 개수 구하기
print(sum(model.pvalues >= 0.05))
```

2. 유의확률이 0.05 미만인 유의한 변수만을 사용해 다시 로지스틱 회귀 분석을 수행하시오. 이 회귀식의 유의한 회귀 계수(상수항 포함)의 합계를 구하시오. (반올림하여 소수 셋째 자리까지 계산)

```python
model.pvalues < 0.05
```

```python
model2 = logit('Churn ~ DataUsage + DayMins', data=df).fit()
print(model2.summary())
```

```python
# 유의한 회귀계수 합 구하기
print(round(model2.params[model2.pvalues<0.05].sum(), 3))
```

1. 주어진 데이터를 이용해 종속변수(PIQ)와 독립변수(Brain, Height, Weight)로 다중 선형 회귀 분석을 수행하시오. 이때 유의확률이 가장 작은 변수의 회귀계수 값을 구하시오. (반올림하여 소수 셋째 자리까지 계산)

```python
import pandas as pd
df = pd.read_csv('/Users/iseonjae/Desktop/빅데이터분석기사/8회기출/piq.csv')
```

```python
print(df.shape)
print(df.head())
print(df.info())
```

```python
# 다중 선형 회귀
from statsmodels.formula.api import ols
model = ols('PIQ ~ Brain + Height + Weight', data=df).fit()
print(model.summary())
```

```python
# 유의확률이 가장 작은 변수의 회귀계수 구하기
model.pvalues.idxmin()
print(round(model.params['Brain'], 3))
```

2. 1번에서 적합한 모델의 결정 계수 값을 구하시오. (반올림하여 소수 둘째 자리까지 계산)

```python
print(round(model.rsquared, 2))
```
